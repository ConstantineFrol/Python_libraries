### logarithmic Diagramm

```python
planets = ['Mercury', 'Venus', 'Earth', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune']
masses = [0.055274, 0.815, 1.0, 0.107, 317.8, 95.0, 14.6, 17.147]
```

```python
# For example Linear Diagramm
plt.plot(planets, masses)
plt.show()
```

```python
masses[planets.index('Earth')] / masses[planets.index('Mercury')]
```

```python
# Logorithmic Vertical diagramm
```

```python
plt.plot(planets, masses)
plt.yscale(value = 'log')
plt.show()
```

```python
planet_info = pd.Series(masses, index = planets)
planet_info
```

# Get Data From URL And Work With It

```python
# Get data from Excel
url = 'http://infographics.economist.com/2018/databank/BMFile2000toJan2018.xls'
```

```python
bmi = pd.read_excel(url)
```

```python
# Check Shape of DataFrame (57 rows, 11 Columns)
bmi.shape
```

### Display names of all columns

```python
bmi.columns
```

### Show Num Of Columns

```python
len(bmi.columns)
```

### Display data to work with

```python
bmi.head()
```

```python
# Display using iloc[]
# [rows:columns].fromTop(rows)
bmi.iloc[:,:4].head(5)
```

### Create Series

```python
# Series = 2 Columns: id and dataColumn
# In this case set 1st column 'Country' as id, 2nd column will be 'dollar_price'
bm_price_series = bmi.set_index('Country')['dollar_price']
```

```python
# Display 5 top rows
bm_price_series.head()
```

### Create Horizontal Graph

```python
# Create graph
bm_price_series.plot(kind = 'barh', logx = True)
# Show graph
plt.show()
```

### Change Size

```python
# from pylab import rcParams
# width, height
rcParams['figure.figsize'] = 9, 20
```

```python
# Create graph
bm_price_series.plot(kind = 'barh', logx = True)
# Show graph
plt.show()
```

## Sort

```python
# Country names is 'id' column, graph is the second column 'dollar_price' (contains values)
# Sort by values( sort by 2nd column )
bm_price_series.sort_values().plot(kind = 'barh', logx = True)
plt.show()
```

```python
# Country names is 'id' column, graph is the second column 'dollar_price' (contains values)
# Sort by index( sort by 1st column )
# RESULT ALWAYS Z-A
bm_price_series.sort_index().plot(kind = 'barh', logx = True)
plt.show()
```

#### Sort A - Z

```python
# Country names is 'id' column, graph is the second column 'dollar_price' (contains values)
# Sort by values( sort by 2nd column )
bm_price_series.sort_index(ascending=False).plot(kind = 'barh', logx = True)
plt.show()
```

### Vertical Graph

```python
rcParams['figure.figsize'] = 4.5, 6
plt.hist(bm_price_series, ec = 'white', color = 'lightgrey')
plt.show()
```
