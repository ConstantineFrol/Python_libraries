# Tuning SVC model

c_vals = 1, pick random numbers 29 times from -2 to 5

```python
c_vals = np.logspace(-2, 5, 29)
c_vals
```

Now collect some data about training and validating

```python
accuracy_valid = []
accuracy_train = []
for val in c_vals:
    clf = SVC(C = val)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_valid)
    y_pred_train = clf.predict(X_train)
    acc_valid = accuracy_score(y_valid, y_pred)
    acc_train = accuracy_score(y_train, y_pred_train)
    accuracy_valid.append(acc_valid)
    accuracy_train.append(acc_train)
    print('C = {} \n\tacc_valid = {} \n\tacc_train = {}\n'.format(val, acc_valid, acc_train))
```

# Show result in graph

```python
import matplotlib.pyplot as plt
%config InlineBackend.figure_format = 'svg'
%matplotlib inline
```

```python
plt.plot(c_vals, accuracy_valid)
plt.plot(c_vals, accuracy_train)
plt.xlabel('C value graph')
plt.ylabel('Accuracy')
plt.legend(['valid', 'train'])
```
