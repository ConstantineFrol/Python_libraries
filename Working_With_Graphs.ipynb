# Import all libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import rcParams
```

### Set Graph size

```python
%matplotlib inline
rcParams['figure.figsize'] = 6, 4.5
```

# Prepare data to use it in graph

```python
# Create np.array, 101 digit from -5 to 5
x = np.linspace(-5, 5, 101)
x
```

```python
# set each value into the power of 2, store in 'x2'
x2 = x**2
x2
```

# Create graph base on X and X2

```python
# Hor = x
# Ver = x2
plt.plot(x, x2)
plt.show()
```

# Add Title

```python
plt.plot(x, x2)
plt.title('Graph f(x)')
plt.show()
```

# Customize fonts

```python
plt.plot(x, x2)
plt.title('Graph f(x)', fontsize = 16)
plt.show()
```

### Fontweight

```python
plt.plot(x, x2)
plt.title('Graph f(x)', fontsize = 16, fontweight = 'bold')
plt.show()
```

### Fontcolor

Version: Name

```python
# Use Colors From HTML table
plt.plot(x, x2)
plt.title('Graph f(x)', fontsize = 16, fontweight = 'bold', color = 'DarkRed')
plt.show()
```

Version: Hex

```python
plt.plot(x, x2)
plt.title('Graph f(x)', fontsize = 16, fontweight = 'bold', color = '#808080')
plt.show()
```

Version: Dict

```python
title_dict = {'fontsize': 20, 'fontweight': 'bold', 'color': '#808080', 'family': 'Monaco'}
```

```python
plt.plot(x, x2)
plt.title('Graph f(x)', fontdict = title_dict)
plt.show()
```

### Title Position

```python
# Default loc = Center
plt.plot(x, x2)
plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')
plt.show()
```

# Axis titles

```python
plt.plot(x, x2)
plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X')

# Axis y (name)[Ver]
plt.ylabel('Function f(x)')

plt.show()
```

### Axis titles Customize

### Display graphs in closer view

Use Axis

```python
# Set Axis size [X_from, X_to, Y_from, Y_to]
plt.axis([-5, 5, -8, 8]) 

plt.plot(x, x2)
plt.plot(x, x3)

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

plt.show()
```

# Add Legend

Version_1

```python
plt.axis([-5, 5, -8, 8]) 

plt.plot(x, x2)
plt.plot(x, x3)

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

plt.legend(labels = ['x2', 'x3'])

plt.show()
```

Version_2

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2')
plt.plot(x, x3, label = 'x**2')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names
plt.legend()

plt.show()
```

### Legend Customize

#### Legend position

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2')
plt.plot(x, x3, label = 'x**2')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right')

plt.show()
```

#### legend fontsize

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2')
plt.plot(x, x3, label = 'x**2')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right', fontsize = 8)

plt.show()
```

#### Legend  use font_dict

```python
legend_dict = {'family': 'Sans', 'size': 10}
```

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2')
plt.plot(x, x3, label = 'x**2')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right', prop = legend_dict)

plt.show()
```

### Legend Frame True, False

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2')
plt.plot(x, x3, label = 'x**2')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right', prop = legend_dict, frameon = False)    

plt.show()
```

# Change Line Color

```python
plt.axis([-5, 5, -8, 8]) 

# Add labels names
plt.plot(x, x2, label = 'x**2', color = '#3bb08f')
plt.plot(x, x3, label = 'x**2', color = '#e97451')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right', prop = legend_dict, frameon = False)    

plt.show()
```

# Change Line Style

```python
plt.axis([-5, 5, -8, 8]) 

# linestyle = 'dashed'
plt.plot(x, x2, label = 'x**2', color = '#3bb08f', linestyle = 'dashed')
# , linestyle = 'dotted'
plt.plot(x, x3, label = 'x**2', color = '#e97451', linestyle = 'dotted')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

# Axis x (name)[Hor]
plt.xlabel('Variable X', fontdict = title_dict)

# Axis y (name)[Ver]
plt.ylabel('Function f(x)', fontdict = title_dict)

# display names in Specific location
plt.legend(loc = 'lower right', prop = legend_dict, frameon = False)    

plt.show()
```

# Add Grid

#### square grid

```python
plt.axis([-5, 5, -8, 8]) 

plt.plot(x, x2, label = 'x**2', color = '#3bb08f', linestyle = 'dashed')

plt.plot(x, x3, label = 'x**2', color = '#e97451', linestyle = 'dotted')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

plt.xlabel('Variable X', fontdict = title_dict)

plt.ylabel('Function f(x)', fontdict = title_dict)

plt.legend(loc = 'lower right', prop = legend_dict, frameon = False)    

# Add grid
plt.grid()

plt.show()
```

#### line grid

```python
plt.axis([-5, 5, -8, 8]) 

plt.plot(x, x2, label = 'x**2', color = '#3bb08f', linestyle = 'dashed')

plt.plot(x, x3, label = 'x**2', color = '#e97451', linestyle = 'dotted')

plt.title('Graph f(x)', fontdict = title_dict, loc = 'left')

plt.xlabel('Variable X', fontdict = title_dict)

plt.ylabel('Function f(x)', fontdict = title_dict)

plt.legend(loc = 'lower right', prop = legend_dict, frameon = False)    

# Add line grid
plt.grid(axis = 'y')

plt.show()
```
