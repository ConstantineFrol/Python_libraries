```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

# ***Figure*** and ***axes*** are base objects in Matplotlib

Create Graph

```python
x = np.linspace(-5, 5, 101)
x_2 = x**2
x_3 = x**3
```

```python
plt.axis([-5, 5, -8, 8])
plt.plot(x,x_2)
plt.plot(x,x_3)
```

# Create Axes and put into row_1 and Column_1 -> graph: x_2

```python
fig, ax = plt.subplots(nrows = 1, ncols = 1)
ax.plot(x, x_2)
```

# Create Axes with 1 row and  2 Columns

### Axes row_1, Col_1 => x_2, row_1, Col_2 => x_3

Version_1 

```python
# In other words: split one graph into induvidual graphs
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax_1, ax_2 = ax
ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
```

Version_2 a_better_way

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax_1, ax_2 = ax.flatten()
ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
```

# object ***Figure*** and it's ***properties***

### Set Figure size which contains axes(graphs)

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax_1, ax_2 = ax.flatten()
ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
fig.set_size_inches(8, 3.5)
```

### Add extra width between Graphs

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
ax_1, ax_2 = ax.flatten()
ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
fig.set_size_inches(8, 3.5)
plt.subplots_adjust(wspace = 0.3)
```

# Create Axes with 2 rows and  1 Column

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)

# Display 2D shape (rows_no, Col_no)
print(ax.shape)

ax_1, ax_2 = ax.flatten()
ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
fig.set_size_inches(3, 6)
```

### Add extra height between Graphs

```python
fig, ax = plt.subplots(nrows = 2, ncols = 1)
ax_1, ax_2 = ax.flatten()

ax_1.plot(x,x_2)
ax_2.plot(x,x_3)

fig.set_size_inches(3, 6)

# Add extra space (percent) '0.25 = 25%'
plt.subplots_adjust(hspace = 0.25)
```

# Create Axes with 2 rows and  2 Columns

### Add extra 2 graphs

```python
z = np.cos(x)
y = np.sin(x)
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,x_2)
ax_2.plot(x,x_3)
ax_3.plot(x,y)
ax_4.plot(x,z)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```

### Set ***Title*** for each graph

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,x_2)
ax_1.set_title('Graph x**2')

ax_2.plot(x,x_3)
ax_2.set_title('Graph x**3')

ax_3.plot(x,y)
ax_3.set_title('Graph Sin(x)')

ax_4.plot(x,z)
ax_4.set_title('Graph Cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```

### Set axis for X and Y width same properties

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, sharey = True)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,x_2)
ax_1.set_title('Graph x**2')

ax_2.plot(x,x_3)
ax_2.set_title('Graph x**3')

ax_3.plot(x,y)
ax_3.set_title('Graph Sin(x)')

ax_4.plot(x,z)
ax_4.set_title('Graph Cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```

### Set by X axis 'graph x**2'

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, sharey = True)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,x_2)
ax_1.set_title('Graph x**2')
ax_1.set_xlim([-3, 3])


ax_2.plot(x,x_3)
ax_2.set_title('Graph x**3')

ax_3.plot(x,y)
ax_3.set_title('Graph Sin(x)')

ax_4.plot(x,z)
ax_4.set_title('Graph Cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```

### Set by Y axis 'graph x**3'

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2, sharey = True)
ax_1, ax_2, ax_3, ax_4 = ax.flatten()

ax_1.plot(x,x_2)
ax_1.set_title('Graph x**2')


ax_2.plot(x,x_3)
ax_2.set_title('Graph x**3')
ax_2.set_ylim([-25, 25])

ax_3.plot(x,y)
ax_3.set_title('Graph Sin(x)')

ax_4.plot(x,z)
ax_4.set_title('Graph Cos(x)')

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.5, hspace = 0.5)
```
